###Gapminder Dataset: https://github.com/chendaniely/pandas_for_everyone/blob/master/data/gapminder.tsv

```python
# Navpreet Kloy, Milan

import pandas as pd
import numpy as np

pd.read_csv('gapminder.tsv', delimiter='\t')  # reads the gapminder file
df = pd.read_csv('gapminder.tsv', delimiter='\t') # gapminder file read and stored in df

print(df.isnull().sum())  # Counts the number of missing (NaN) values in each column

```

```python
# Standardize selected columns
from sklearn import preprocessing #import tool to standardize scale

scale = ['lifeExp', 'pop', 'gdpPercap'] # columns to be scaled from gapminder stored into variable

df[scale] = preprocessing.scale(df[scale])  # selected columns scaled and replaced in df DataFrame

df #displays updated DataFrame
```

```python
# Reording column headers
old_cols=df.columns.values # coverts and stores the column index to a NumPy array
new_cols=['year','country','continent',	'pop','lifeExp','gdpPercap'] # creates new list with desired reorder
df= df.reindex(columns=new_cols) # reorders the DataFrame columns as specified in new_cols
df #displays the DataFrame
```

```python
# Renaming column headers
print(df.rename(columns={'pop':'population'}))  # prints DataFrame with renamed column
```

```python
# Converting String to Numeric Values
from sklearn.preprocessing import LabelEncoder #LabelEncoder imported from sklearn.preprocessing

le = LabelEncoder()  # Instance of LabelEncoder created

label = le.fit_transform(df['continent']) #LabelEncoder finds unique continent values and transforms them into integers, which are saved into label

df.drop("continent",axis=1, inplace=True)  # removes the original 'continent' column
df["continent"] = label # appends the encoded label with columns name 'continent
df #displays DataFrame, in this showcasing the updated encoded continent column

```

###Tesla Stock Dataset: https://github.com/chendaniely/pandas_for_everyone/blob/master/data/tesla_stock_yahoo.csv


```python

tesla_stock_yahoo_df = pd.read_csv('tesla_stock_yahoo.csv', delimiter=',')  # tesla stock file read and stored in tesla_stock_yahoo_df

print(tesla_stock_yahoo_df.isnull().sum())  # Counts the number of missing (NaN) values in each column

tesla_stock_yahoo_df.head() #prints first five rows of DataFrame


```

```python
# Standardize selected columns
from sklearn import preprocessing #import tool to standardize scale

scale = ['Open', 'High', 'Low','Close','Adj Close','Volume'] # columns to be scaled from gapminder stored into variable


tesla_stock_yahoo_df[scale] = preprocessing.scale(tesla_stock_yahoo_df[scale])  # selected columns scaled and replaced in df DataFrame

tesla_stock_yahoo_df #displays updated DataFrame
```

```python
# Reording column headers
old_cols=tesla_stock_yahoo_df.columns.values # coverts and stores the column index to a NumPy array
new_cols=['Date','Low', 'High', 'Open','Close','Adj Close','Volume'] # creates new list with desired reorder
tesla_stock_yahoo_df= tesla_stock_yahoo_df.reindex(columns=new_cols) # reorders the DataFrame columns as specified in new_cols
tesla_stock_yahoo_df #displays the DataFrame
```

```python
# Renaming column headers
print(tesla_stock_yahoo_df.rename(columns={'Adj Close':'Adjusted Closing', 'Volume':'Vol'}))  # prints DataFrame with renamed column
```

```python
# Converting String to Numeric Values
from sklearn.preprocessing import LabelEncoder #LabelEncoder imported from sklearn.preprocessing

le = LabelEncoder()  # Instance of LabelEncoder created

label = le.fit_transform(tesla_stock_yahoo_df['Date']) #LabelEncoder finds unique Date values and transforms them into integers, which are saved into label

tesla_stock_yahoo_df.drop("Date",axis=1, inplace=True)  # removes the original 'Date' column
tesla_stock_yahoo_df["Date"] = label # appends the encoded label to Date column
tesla_stock_yahoo_df #displays DataFrame with Date column encoded
```
